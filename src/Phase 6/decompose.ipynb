```python
import pandas as pd
from pandas import read_csv
from datetime import datetime
```

```python
date_parser = lambda d: datetime.strptime(d, '%Y%m')

# Read Dataset from CSV into Pandas Dataframe.
data_df = read_csv(
    '../../data/data.csv',
    names=['date', 'temperature'],
    dtype={'date': 'str', 'temperature': 'float'},
    parse_dates=['date'],
    date_parser=date_parser,
    index_col='date'
)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib

decompose_res = seasonal_decompose(data_df, model='additive', freq=12)
matplotlib.rcParams.update({'font.size': 26, 'figure.figsize': '45, 25'})
fig = decompose_res.plot()
fig.suptitle('Observed Climate Data')
```
